```python
import csv
import re

isNumPattern = re.compile("\d{1,}\.*\d{0,}")

dataList = list()

# mo file va thuc hien doc du lieu, ket hop
with open('data.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        if (dataList == []):
            for i in range(0, row.__len__()):
                dataList.append(list())
        for i in range(0, row.__len__()):
            try:
                dataList[i].append(float(row[i]))
            except:
                dataList[i].append(row[i])
# for rowLst in dataList:
print(dataList)
```
